```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/Diwali Sales Data.csv', encoding='unicode_escape') # to avoid uncode error write unicode_escape
data.head()
```

```python
data.shape
```

```python
data.info()# in these data base clearly see status and unnamed1 coloumn having no data
```

```python
#remove blank column
data.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
data.info()# as you see two column are removed
```

```python
pd.isnull(data).sum()# you can see 12 null value are availble
```

```python
data.shape
```

```python
data.dropna(inplace = True) # drop null value
```

```python
data.shape
```

```python
data['Amount'] = data['Amount'].astype('int')
```

```python
data['Amount'].dtypes
```

```python
data.columns
```

```python
data.rename(columns={'Marital_Status':'shadi'})
```

```python
#Exploratory Data Analysis
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ax = sns.countplot(x = 'Gender', data= data, color= 'lightgray')
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
data.columns
```

```python
sns.countplot(x = 'Age Group', data = data, hue = 'Gender', color = 'lightgray')
```

```python
ax = sns.countplot(x= 'Age Group', data = data, hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#total amount vs Age group
sales_age = data.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

ax=sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

# **State**

```python
data.columns
```

```python
sales_state = data.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(5)
sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
sales_state = data.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state, color = 'orange')
```

#Marital Status

```python
data.columns
```

```python
ax = sns.countplot(x='Marital_Status', data= data )
sns.set(rc= {'figure.figsize':(5,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Mstatus = data.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(5)
sns.set(rc= {'figure.figsize':(5,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = Mstatus, hue = 'Gender')
```

```python
data.columns
```

```python
data_sorted = data.sort_values(by='Occupation', ascending=True)
sns.set(rc={'figure.figsize':(22,5)})
ax = sns.countplot(x = 'Occupation', data = data_sorted, order=data_sorted['Occupation'].value_counts().index)
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
sales_ocupation =  data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_ocupation, x = 'Occupation',y= 'Amount')
```

#Product catagory

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = data, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = data.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
